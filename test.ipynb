```python
numeral_dict = {
    1000: 'M',
    900: 'CM',
    500: 'D',
    400: 'CD',
    100: 'C',
    90: 'XC',
    50: 'L',
    40: 'XL',
    10: 'X',
    9: 'IX',
    5: 'V',
    4: 'IV',
    1: 'I'
}
```

Algorithm:
1. go from largest to smallest over dict of roman numerals until it is less than or equal to zero
2. do // division and do: number - (roman num value * div result)

```python
roman_result = ''
def numeral_converter(number_input):
    roman_result = ''
    while number_input > 0:
        for arabic_num, roman_num in numeral_dict.items():
            if number_input >= arabic_num:
                floor_result = number_input // arabic_num
                number_input = number_input - (floor_result * arabic_num)
                roman_result = roman_result + (roman_num * floor_result)
                print(f"roman_result: {roman_result}")
    return roman_result
```

```python
result = numeral_converter(378)
result
```
